```python
from collections import Counter 
```

```python
columns = [ 'age','year','nodes','class']
data_names =  pd.read_csv('haberman.csv',header = None,names=columns)
target = data_names['class'].values
counter =Counter(target)
data_names
```

```python
for key,value in counter.items():
     precentage = value / len(target) * 100
     print('class= %d, Count= %d , precentage = %.3f %%' % (key,value,precentage))
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

x,y = data_names.values[:,:-1],data_names.values[:,-1]
x=x.astype('float32')
y = LabelEncoder().fit_transform(y)
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.9,stratify=y,random_state=3)
```

```python
from tensorflow.keras import Sequential 
from tensorflow.keras.layers import Dense 
n_featuers = x.shape[1]
model = Sequential()
model.add(Dense(10,activation='relu',input_shape=(n_featuers,) ))
model.add(Dense(1,activation='sigmoid' ))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(x_train,y_train,epochs=200,batch_size=20,validation_data=(x_test,y_test))
```

```python
from sklearn.metrics import accuracy_score

y_predict = (model.predict(x_test) >0.5).astype(int)
score= accuracy_score(y_test,y_predict)
score
```

```python
from matplotlib import pyplot as pt
pt.title('learning Curve ::lose')
pt.xlabel('Epoch',fontsize=14)
pt.ylabel('Cross Entropy',fontsize=14)
pt.plot(history.history['loss'],label='train')
pt.plot(history.history['val_loss'],label='validation')
pt.legend()
pt.show()
```
